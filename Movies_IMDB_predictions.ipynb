```python
import pandas as pd
import seaborn as sns
```

```python
movies = pd.read_csv("/content/movies.csv")
movies.head()
```

```python
movies.tail()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
catcol = movies[['title','year','certificate','genre','director','stars']]

for i in catcol:
  movies[i]=movies[i].astype('category')
```

```python
movies.info()
```

```python
print(movies[['certificate']].value_counts())
```

```python
movies[['certificate']]=movies[['certificate']].fillna('R')
```

```python
movies.isnull().sum()
```

# Checking skewness:


```python
print(movies[['metascore']].skew())  # normal--mean
```

```python
print(movies[['gross']].skew())     # skewed - median
```

```python
movies['metascore']=movies['metascore'].fillna(movies['metascore'].mean())
```

```python
movies['gross']=movies['gross'].fillna(movies['gross'].median())
```

```python
movies.isnull().sum()
```

```python
# checking correlation
matrix =movies.corr()
```

```python
sns.heatmap(matrix,annot=True)
```

#Defining x and y:


```python
x=movies[['runtime','rating','metascore','votes']]
y=movies[['gross']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
```

#Splitting the data into Train and Test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

**Linear Regression**:

```python
lr = LinearRegression()
model =lr.fit(x_train,y_train)
```

```python
y_test['Predictions_using_LinearRegression']=model.predict(x_test)
y_test
```

```python
model.score(x_train,y_train)
```

**Ridge Regression**

```python
ridge = Ridge()
model1 = ridge.fit(x_train,y_train)
model1.score(x_train,y_train)
```

**Lasso Regression**:

```python
lasso=Lasso()
model2=lasso.fit(x_train,y_train)
model2.score(x_train,y_train)
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score
```

```python

random_forest = RandomForestRegressor(n_estimators=200,random_state=42)
random_forest.fit(x_train,y_train)
```

```python
y_test['Random_f_predictions']=random_forest.predict(x_test)
y_test
```

```python
random_forest.score(x_train,y_train)
```

```python
rf=RandomForestRegressor(random_state=42)
from pprint import pprint

# current parameters-by default
print('Params currently in use:\n')
pprint(rf.get_params())
```

# Using K-FOLD cross validation

```python
# using kfold
from sklearn.model_selection import KFold,cross_val_score
from sklearn.metrics import r2_score
```

```python
# K-FOLD without parameter tuning:

kfold = KFold(n_splits=5,random_state=42,shuffle=True)
result = cross_val_score(random_forest,x,y,cv=kfold,scoring='r2')
print(result.mean())
```

```python
# K-FOLD after parameter tuning:

from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
kfold = KFold(n_splits=5, random_state=23, shuffle=True)
params = {'n_estimators':[25,50,100,150],
          'max_features':['sqrt','log2',None],
          'max_depth':[3,6,9],
          'max_leaf_nodes':[3,6,9]}
random_forest = RandomForestRegressor(n_estimators=200,max_features=2,max_depth=4)
gcv = GridSearchCV(random_forest,param_grid=params,cv=kfold,scoring='r2')
gcv.fit(x,y)
print(gcv.best_score_)
print(gcv.best_params_)
```

```python
# using best parameters obtained above for random_forest:

model_random = RandomForestRegressor(max_depth=9, max_features= None, max_leaf_nodes= 9, n_estimators=100)
model_random.fit(x_train,y_train)
y_pred_random = model_random.predict(x_test)
```

```python
y_test['random_parameter_tuning']=model_random.predict(x_test)
y_test
```

```python
model_random.score(x_train,y_train)
```

```python
y_test
```

#Result:

```python
print("Random_forest_before_tuning:",random_forest.score(x_train,y_train))
print("Random_forest_after_tuning:",model_random.score(x_train,y_train))
print("Linear_Regression:",lr.score(x_train,y_train))
print("Ridge_Regression:",model1.score(x_train,y_train))
print("Lasso_Regression:",model2.score(x_train,y_train))
```
